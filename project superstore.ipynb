```python
#drop unrelated/blank colums
df.drop(columns=['ind1', 'ind2'], inplace=True)
```

```python
# Handle missing values in the 'Returns' column
df['Returns'].fillna(0, inplace=True)
```

```python
# Convert date columns to datetime format with the correct format string
df['Order Date'] = pd.to_datetime(df['Order Date'], format='%d-%m-%Y')
df['Ship Date'] = pd.to_datetime(df['Ship Date'], format='%d-%m-%Y')
```

```python
df.hist(figsize=(15, 10))
plt.tight_layout()
plt.show()
```

```python
# Bar chart for categorical variables
plt.figure(figsize=(10, 6))
df['Category'].value_counts().plot(kind='bar', color='skyblue')
plt.title('Distribution of Categories')
plt.xlabel('Category')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python

# Count the occurrences of each sub-category
sub_category_counts = df['Sub-Category'].value_counts()

# Plot the distribution of sub-categories
plt.figure(figsize=(10, 6))
sub_category_counts.plot(kind='bar', color='skyblue')
plt.title('Distribution of Sub-Categories')
plt.xlabel('Sub-Category')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
payment_mode_counts = df['Payment Mode'].value_counts()

# Create a pie chart
plt.figure(figsize=(6,6))
plt.pie(payment_mode_counts, labels=payment_mode_counts.index, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Payment Modes')
plt.axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
plt.show()
```

```python
segment_sales = df.groupby('Segment')['Sales'].sum()


plt.figure(figsize=(10, 6))
segment_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Segment')
plt.xlabel('Segment')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

```python

# Group the data by state and calculate total sales for each state
state_sales = df.groupby('State')['Sales'].sum().sort_values(ascending=False)

# Select the top 10 states
top_10_states = state_sales.head(10)

# Create a bar plot
plt.figure(figsize=(10, 6))
top_10_states.plot(kind='bar', color='skyblue')
plt.title('Top 10 States by Sales')
plt.xlabel('State')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python

# Group the data by city and calculate total sales for each city
city_sales = df.groupby('City')['Sales'].sum().sort_values(ascending=False)

# Select the top 10 cities
top_10_cities = city_sales.head(10)

# Create a bar plot
plt.figure(figsize=(12, 6))
top_10_cities.plot(kind='bar', color='skyblue')
plt.title('Top 10 Cities by Sales')
plt.xlabel('City')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```

```python
ship_mode_sales = df.groupby('Ship Mode')['Sales'].sum().sort_values(ascending=False)

# Create a bar plot
plt.figure(figsize=(8, 6))
ship_mode_sales.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Ship Mode')
plt.xlabel('Ship Mode')
plt.ylabel('Total Sales')
plt.xticks(rotation=0)
plt.show()
```

```python
customer_sales_profit = df.groupby('Customer Name').agg({'Sales': 'sum', 'Profit': 'sum'})

# Select the top 10 customers based on sales or profit
top_10_customers_sales = customer_sales_profit['Sales'].nlargest(10)
top_10_customers_profit = customer_sales_profit['Profit'].nlargest(10)

# Create visualizations
plt.figure(figsize=(12, 6))

# Plot total sales for the top 10 customers
plt.subplot(1, 2, 1)
top_10_customers_sales.plot(kind='bar', color='skyblue')
plt.title('Top 10 Customers by Sales')
plt.xlabel('Customer Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)

# Plot total profit for the top 10 customers
plt.subplot(1, 2, 2)
top_10_customers_profit.plot(kind='bar', color='orange')
plt.title('Top 10 Customers by Profit')
plt.xlabel('Customer Name')
plt.ylabel('Total Profit')
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

```python
# Group the data by product name and calculate total sales for each product
product_sales = df.groupby('Product Name')['Sales'].sum().sort_values(ascending=False)

# Select the top 10 products
top_10_products = product_sales.head(10)

# Create a bar plot
plt.figure(figsize=(12, 6))
top_10_products.plot(kind='bar', color='orange')
plt.title('Top 10 Products by Sales')
plt.xlabel('Product Name')
plt.ylabel('Total Sales')
plt.xticks(rotation=45)
plt.show()
```
